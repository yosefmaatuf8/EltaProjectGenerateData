```python
import uuid

random_uuid = str(uuid.uuid4())
print(random_uuid)

rule_instance_to_predict = f"""
Please create a rule Type "Covert Agent" , The rule types includes following parameters: id, status, site, importance, gender, activity, origin and cost.
The conditions relate to Agent message.
The Relations between the conditions are "And".

- Id equal to id parameter.
- Status equal to status parameter
- Site equal to site parameter.
- Importance greater then importance parameter.
- Gender equal to gender parameter.
- Activity equal to activity parameters.
- Origin equal to origin parameter.
- Cost greater then cost parameters.
"_id": {random_uuid},

example values:
{{
    "id": string
    "status": string,
    "site": string,
    "importance": Int,
    "gender": string
    "activity": string,
    "origin": string,
    "cost": Int,
    }}"""
```

```python
rule_instance_to_predict
```

```python
template = """
here its an example with input, and what should be the output: input: {rule_instance_description}, example values: {rule_instance_example}, output: {rule_type_example}.
create output for this input: {to_predict}
"""
```

```python
template_params = {"rule_instance_description": rule_instance_description,
                   "rule_instance_example": rule_instance_example,
                   "rule_type_example": rule_type_example,
                   "to_predict": rule_instance_to_predict}
```

```python
prompt = ChatPromptTemplate.from_template(template)
chain = prompt | llm
response = chain.invoke(template_params).content
```
